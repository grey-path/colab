## Задание 1

**A. Импортируйте библиотеку Pandas и дайте ей псевдоним pd.**

```python
import pandas as pd
import numpy as np
```

**B.Создайте датафрейм authors со столбцами author_id и author_name, в которых соответственно содержатся данные: [1, 2, 3] и  ['William Shakespeare', 'Anton Chehov', 'Lev Tolstoy'].**

```python
authors = pd.DataFrame({'author_id':[1, 2, 3],
                        'author_name': ['William Shakespeare', 'Anton Chehov', 'Lev Tolstoy']},
                       columns=['author_id', 'author_name'])
print(authors)
```

**C. C. Затем создайте датафрейм book cо столбцами author_id, book_title и price,в которых соответственно содержатся данные: [1, 1, 1, 2, 2, 3, 3],['Гамлет', 'Король Лир', 'Ромео и Джульетта', 'Толстый и тонкий', 'Дама с собачкой', ' Война и мир', 'Детство'], [450, 300, 350, 500, 450, 370, 290].**

```python
book = pd.DataFrame({'author_id':[1, 1, 1, 2, 2, 3, 3],
                     'book_title':['Гамлет', 'Король Лир', 'Ромео и Джульетта', 'Толстый и тонкий', 'Дама с собачкой', ' Война и мир', 'Детство'],
                     'price':[450, 300, 350, 500, 450, 370, 290]},
                    columns=['author_id', 'book_title', 'price'])
print(book)
```

## Задание 2

**Получите датафрейм authors_price, соединив датафреймы authors и books по полю author_id.**

```python
authors_price = pd.merge(authors, book)
print(authors_price)
```

## Задание 3

**Создайте датафрейм top5, в котором содержатся строки из authors_price с пятью самыми дорогими книгами.**

```python
top5 = authors_price.nlargest(5, 'price')
print(top5)
```

```python
top5 = authors_price.sort_values(by = 'price', ascending = False)
top5[:5]
```

## Задание 4

**A. Создайте датафрейм authors_stat на основе информации из authors_price.**

```python
authors_stat = authors_price['author_name'].value_counts()
print(authors_stat)
```

**B. В датафрейме authors_stat должны быть четыре столбца:
author_name, min_price, max_price и mean_price, в которых должны содержаться соответственно имя автора, минимальная, максимальная и средняя цена на книги этого автора.**

```python
authors_stat = authors_price.groupby('author_name').agg({'price':['min', 'max', 'mean']})
authors_stat = authors_stat.rename(columns={'min':'min_price', 'max':'max_price', 'mean':'mean_price'})
print(authors_stat)
```

```python
pivot = pd.pivot_table(
    data=authors_price,
    index='author_name',

    aggfunc={'price':['min', 'max','mean']}
)
pivot
```

```python
authors_stat= authors_price.groupby('author_name').agg(min_price=('price', 'min'),
max_price=('price', 'max'), mean_price=('price', 'mean')).round(2)
authors_stat
```

**Создайте новый столбец в датафрейме authors_price под названием cover, в нем будут располагаться данные о том, какая обложка у данной книги - твердая или мягкая. В этот столбец поместите данные из следующего списка: ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая'].

Просмотрите документацию по функции pd.pivot_table с помощью вопросительного знака. Для каждого автора посчитайте суммарную стоимость книг в твердой и мягкой обложке. Используйте для этого функцию pd.pivot_table, crosstable и  groupby. При этом столбцы должны называться "твердая" и "мягкая", а индексами должны быть фамилии авторов. Пропущенные значения стоимостей заполните нулями, при необходимости загрузите библиотеку Numpy. Назовите полученный датасет book_info и сохраните его в формат csv под названием "book_info.csv".**

```python
authors_price['cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
print(authors_price)
```

```python
book_info = pd.pivot_table(authors_price, values='price', index=['author_name'], columns=['cover'], aggfunc=np.sum)
book_info['мягкая'] = book_info['мягкая'].fillna(0)
book_info['твердая'] = book_info['твердая'].fillna(0)
print(book_info)
```

```python
import numpy as np
import pandas as pd
fruit = pd.Series(np.random.choice(['apple', 'banana', 'carrot'], 10))
weights = pd.Series(np.linspace(1, 10, 10))
#solution
dataFrame = pd.DataFrame({'fruit': fruit,'weights': weights})
print(dataFrame)
print()
meanA = np.mean((dataFrame.loc[dataFrame['fruit']=='apple','weights']))
print("The averge weight of apples is " + str("{:.1f}".format(meanA)))
meanB = np.mean((dataFrame.loc[dataFrame['fruit']=='banana','weights']))
print("The averge weight of bananas is " + str("{:.1f}".format(meanB)))
meanC = np.mean((dataFrame.loc[dataFrame['fruit']=='carrot','weights']))
print("The averge weight of carrots is " + str("{:.1f}".format(meanC)))
```
