```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = x ** 2

plt.plot(x, y)
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

Постройте на одном графике синус и косинус от 0 до 2π. Используйте разные цвета и стили линий.

```python
x = np.linspace(0, 2 * np.pi, 100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='sin(x)', color='blue', linestyle='--')
plt.plot(x, y_cos, label='cos(x)', color='red', linestyle='-')
plt.title('Sine and Cosine Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

Создайте фигуру с двумя графиками: на первом графике покажите экспоненциальную функцию y=e^x , на втором — натуральный логарифм y=ln⁡(x).

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

x1 = np.linspace(0, 2, 100)
y1 = np.exp(x1)
ax1.plot(x1, y1, color='green')
ax1.set_title('y = e^x')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.grid(True)

x2 = np.linspace(0.1, 10, 100)
y2 = np.log(x2)
ax2.plot(x2, y2, color='orange')
ax2.set_title('y = ln(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

Создайте круговую диаграмму, отображающую доли различных категорий (например, фруктов: яблоки, бананы, апельсины, виноград).

```python
labels = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sizes = [25, 35, 20, 20]
colors = ['red', 'yellow', 'orange', 'purple']
explode = (0.1, 0, 0, 0)  # выделить сектор с яблоками

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Fruit Distribution')
plt.show()
```
