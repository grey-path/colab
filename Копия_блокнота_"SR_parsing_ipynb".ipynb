```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.gutenberg.org/ebooks/search/?query=Dostoe&submit_search=Go%21'

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

data = []
for book in soup.find_all('li', class_='booklink'):
    author = book.find('span', class_='subtitle')
    book_name = book.find('span', class_='title')
    downloads = book.find('span', class_='extra')
    author = str(author)[23:-7]
    book_name = str(book_name)[20:-7]
    downloads = str(downloads)[20:-17]
    downloads = int(downloads)
    data.append({'author': author, 'book_name': book_name, 'downloads': downloads})

df = pd.DataFrame(data)
df.head()
```
