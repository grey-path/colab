**8. Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак marital-status)? Женатыми считаем тех, у кого marital-status начинается с Married (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.**


```python
mar_status = [ word for word in df['marital-status'].unique() if word.startswith('Married') ]
mar_status
```

```python
arr_marr = df[(df['sex'] == 'Male') &
     (df['marital-status'].isin(mar_status))]['income'].value_counts()
print(arr_marr)
```

```python
arr_not_marr = df[(df['sex'] == 'Male') &
     (~df['marital-status'].isin(mar_status))]['income'].value_counts()
print(arr_not_marr)
print("Богатых среди женатых {0}%".format(round(100 * arr_marr[1]/(arr_marr[0]+arr_marr[1]), 2)))
print("Богатых среди холостых {0}%".format(round(100 * arr_not_marr[1]/(arr_not_marr[0]+arr_not_marr[1]), 2)))
```

```python
totally_men = len(df[(df['income']=='>50K')][(df['sex']=='Male')])
print('Totally men:', totally_men)

a = len(df[(df['marital-status']=='Married-civ-spouse')][(df['income']=='>50K')][(df['sex']=='Male')])
#print('Married-civ-spouse', a)
b = len(df[(df['marital-status']=='Married-spouse-absent')][(df['income']=='>50K')][(df['sex']=='Male')])
#print('Married-spouse-absent', b)
c = len(df[(df['marital-status']=='Married-AF-spouse')][(df['income']=='>50K')][(df['sex']=='Male')])
#print('Married-AF-spouse', c)
totally_married_men = a + b + c
print('Totally married men >50K:', totally_married_men)
```

**9. Какое максимальное число часов человек работает в неделю (признак hours-per-week)? Сколько людей работают такое количество часов и каков среди них процент зарабатывающих много?**

```python
max_work = df['hoursPerWeek'].max()
print("Макс. число часов - {0} час./нед.".format(max_work))

num_workers = df[df['hoursPerWeek'] == max_work].shape[0]
print("Всего людей с таким кол-ом часов {0}".format(num_workers))

per_rich = df[(df['hoursPerWeek'] == max_work) & (df['income'] == '>50K')].shape[0] / num_workers
print("Из них богатых {0}%".format(round(100 * per_rich, 2)))
```
