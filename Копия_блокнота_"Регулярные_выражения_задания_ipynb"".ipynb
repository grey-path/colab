**–†–µ–≥—É–ª—è—Ä–Ω—ã–µ –≤—ã—Ä–∞–∂–µ–Ω–∏—è** (regular expression, regex, —Ä–µ–≥—É–ª—è—Ä–∫–∞) ‚Äì —ç—Ç–æ –º–æ—â–Ω—ã–π —Å–ø–æ—Å–æ–± –æ–±—Ä–∞–±–æ—Ç–∫–∏ —Ç–µ–∫—Å—Ç–∞, –ø–æ–∑–≤–æ–ª—è—é—â–∏–π –Ω–∞—Ö–æ–¥–∏—Ç—å –≤ —Å—Ç—Ä–æ–∫–∞—Ö –ø–æ–¥—Å—Ç—Ä–æ–∫–∏ –ø–æ —Å–ª–æ–∂–Ω—ã–º —É—Å–ª–æ–≤–∏—è–º.

–ß—É—Ç—å –±–æ–ª–µ–µ —Ñ–æ—Ä–º–∞–ª—å–Ω–æ, —Ä–µ–≥—É–ª—è—Ä–Ω–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ ‚Äî —ç—Ç–æ —Å—Ç—Ä–æ–∫–∞, –∑–∞–¥–∞—é—â–∞—è —à–∞–±–ª–æ–Ω –ø–æ–∏—Å–∫–∞ –ø–æ–¥—Å—Ç—Ä–æ–∫ –≤ —Ç–µ–∫—Å—Ç–µ. –û–¥–Ω–æ–º—É —à–∞–±–ª–æ–Ω—É –º–æ–∂–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –º–Ω–æ–≥–æ —Ä–∞–∑–Ω—ã—Ö —Å—Ç—Ä–æ–∫. –†–µ–≥—É–ª—è—Ä–Ω–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ —Å–æ—Å—Ç–æ–∏—Ç –∏–∑ –æ–±—ã—á–Ω—ã—Ö —Å–∏–º–≤–æ–ª–æ–≤ –∏ —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã—Ö –∫–æ–º–∞–Ω–¥–Ω—ã—Ö –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–µ–π. –ù–∞–ø—Ä–∏–º–µ—Ä, \d –∑–∞–¥–∞—ë—Ç –ª—é–±—É—é —Ü–∏—Ñ—Ä—É, –∞ \d+ ‚Äî –∑–∞–¥–∞–µ—Ç –ª—é–±—É—é –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∏–∑ –æ–¥–Ω–æ–π –∏–ª–∏ –±–æ–ª–µ–µ —Ü–∏—Ñ—Ä.

–†–∞–±–æ—Ç–∞ —Å —Ä–µ–≥—É–ª—è—Ä–Ω—ã–º–∏ –≤—ã—Ä–∞–∂–µ–Ω–∏—è–º–∏ —Ä–µ–∞–ª–∏–∑–æ–≤–∞–Ω–∞ –≤–æ –≤—Å–µ—Ö —Å–æ–≤—Ä–µ–º–µ–Ω–Ω—ã—Ö —è–∑—ã–∫–∞—Ö –ø—Ä–æ–≥—Ä–∞–º–º–∏—Ä–æ–≤–∞–Ω–∏—è. –û–¥–Ω–∞–∫–æ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ –¥–∏–∞–ª–µ–∫—Ç–æ–≤, –ø–æ—ç—Ç–æ–º—É —Ñ—É–Ω–∫—Ü–∏–æ–Ω–∞–ª —Ä–µ–≥—É–ª—è—Ä–Ω—ã—Ö –≤—ã—Ä–∞–∂–µ–Ω–∏–π –º–æ–∂–µ—Ç —Ä–∞–∑–ª–∏—á–∞—Ç—å—Å—è –æ—Ç —è–∑—ã–∫–∞ –∫ —è–∑—ã–∫—É.

üìå–ù–∞–ø—Ä–∏–º–µ—Ä, –º–æ–∂–Ω–æ –ø–æ–Ω—è—Ç—å, –∫–∞–∫–∏–µ –¥–∞—Ç—ã –≤ –Ω–∞–±–æ—Ä–µ —Å—Ç—Ä–æ–∫ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω—ã –≤ —Ñ–æ—Ä–º–∞—Ç–µ DD/MM/YYYY, –∞ –∫–∞–∫–∏–µ - –≤ –¥—Ä—É–≥–∏—Ö —Ñ–æ—Ä–º–∞—Ç–∞—Ö.

–ò–ª–∏ –±—ã–≤–∞–µ—Ç, –Ω–∞–ø—Ä–∏–º–µ—Ä, —á—Ç–æ –ø–µ—Ä–µ–¥ —Ä–∞–±–æ—Ç–æ–π —Å —Ç–µ–∫—Å—Ç–æ–º, –Ω–∞–¥–æ –ø–æ—á–∏—Å—Ç–∏—Ç—å –µ–≥–æ –æ—Ç —Å–≤–æ–µ–æ–±—Ä–∞–∑–Ω–æ–≥–æ –º—É—Å–æ—Ä–∞: —É–ø–æ–º–∏–Ω–∞–Ω–∏–π –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π, url –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ.

–ù–∞–≤—ã–∫ –ø–æ–ª–µ–∑–Ω—ã–π, –¥–∞–≤–∞–π—Ç–µ –≤ –Ω—ë–º —Ç–æ–∂–µ –ø–æ—Ç—Ä–µ–Ω–∏—Ä—É–µ–º—Å—è.

````python
text = "Order ID: 12345, Invoice No: 67890"

# –ù–∞–π–¥—ë–º ID –∑–∞–∫–∞–∑–∞ –∏ —Å—á—ë—Ç–∞
match = re.search(r"Order ID: (\d+), Invoice No: (\d+)", text)

if match:
    print("ID –∑–∞–∫–∞–∑–∞:", match.group(1))
    print("–ù–æ–º–µ—Ä —Å—á—ë—Ç–∞:", match.group(2))
````

````python
text = "USD 100, EUR 200, GBP 300"

# –ù–∞–π–¥—ë–º —Å—É–º–º—ã, –Ω–µ –≤—ã–¥–µ–ª—è—è –≤–∞–ª—é—Ç—É
matches = re.findall(r"(?:USD|EUR|GBP) (\d+)", text)
print("–°—É–º–º—ã:", matches)

# –ù–∞–π–¥—ë–º —Å—É–º–º—ã –∏ –≤–∞–ª—é—Ç—É
matches = re.findall(r"(USD|EUR|GBP) (\d+)", text)
print("–°—É–º–º—ã:", matches)
````

````python
text = "ID12345 is confirmed. ID23456 is confirmed"

# –ü—Ä–æ–≤–µ—Ä—è–µ–º, –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è –ª–∏ —Å—Ç—Ä–æ–∫–∞ —Å "ID" + —Ü–∏—Ñ—Ä—ã
match = re.match(r"ID\d+", text)

if match:
    print("–û–±—ä–µ–∫—Ç Match:", match)
    print("–°–∞–º–æ —Å–æ–≤–ø–∞–¥–µ–Ω–∏–µ:", match.group())
    print("–î–∏–∞–ø–∞–∑–æ–Ω —Å–æ–≤–ø–∞–¥–µ–Ω–∏—è:", match.span())
else:
    print("–ù–µ—Ç —Å–æ–≤–ø–∞–¥–µ–Ω–∏—è.")
````

#**–ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –∑–∞–¥–∞–Ω–∏–µ 1**

````python
import re

def validate_password(password):
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –¥–ª–∏–Ω—ã –ø–∞—Ä–æ–ª—è
    if len(password) < 8:
        return False

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ–π –∑–∞–≥–ª–∞–≤–Ω–æ–π –±—É–∫–≤—ã
    if not re.search(r'[A-Z]', password):
        return False

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ–π —Å—Ç—Ä–æ—á–Ω–æ–π –±—É–∫–≤—ã
    if not re.search(r'[a-z]', password):
        return False

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞–ª–∏—á–∏—è —Ö–æ—Ç—è –±—ã –æ–¥–Ω–æ–π —Ü–∏—Ñ—Ä—ã
    if not re.search(r'[0-9]', password):
        return False

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –¥–æ–ø—É—Å—Ç–∏–º—ã–µ —Å–ø–µ—Ü–∏–∞–ª—å–Ω—ã–µ —Å–∏–º–≤–æ–ª—ã
    if not re.fullmatch(r'[A-Za-z0-9!@#$%^&*()]*', password):
        return False

    return True
````

````python
import re

text = "Server1: 192.168.1.1, Server2: 10.0.0.254, Invalid: 999.123.456.78"

# –†–µ–≥—É–ª—è—Ä–Ω–æ–µ –≤—ã—Ä–∞–∂–µ–Ω–∏–µ –¥–ª—è –ø–æ–∏—Å–∫–∞ IPv4-–∞–¥—Ä–µ—Å–æ–≤
ip_pattern = r'\b(25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.(25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.(25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\.(25[0-5]|2[0-4][0-9]|[01]?[0-9][0-9]?)\b'

# –ù–∞—Ö–æ–¥–∏–º –≤—Å–µ —Å–æ–≤–ø–∞–¥–µ–Ω–∏—è
ip_addresses = re.findall(ip_pattern, text)

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∫–æ—Ä—Ç–µ–∂–∏ –≤ —Å—Ç—Ä–æ–∫–∏ —Å —Ç–æ—á–∫–∞–º–∏
valid_ips = ['.'.join(ip) for ip in ip_addresses]

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
for ip in valid_ips:
    print(ip)
````
