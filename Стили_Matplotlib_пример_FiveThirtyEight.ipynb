```python
style.use('Solarize_Light2')
plt.plot([1, 2, 3], [5, 2, 7])
plt.show()

style.use('default')
plt.plot([1, 2, 3], [5, 2, 7])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9, 5))
# Assume the other code is written
ax.barh(white_wine.corr()['quality'].index, white_wine.corr()['quality'], left=2)
ax.barh(red_wine.corr()['quality'].index, red_wine.corr()['quality'])
```

Основываясь на принципе близости, мы теперь видим два разных графика баров: участок красного вина слева (красный цвет) и участок белого вина справа (синий цвет).

# Добавление подписи
Графики FiveThirtyEight обычно имеют полосу подписи внизу, где мы видим имя создателя и источник данных.


Matplotlib не предоставляет нам функции, предназначенной для создания полос сигнатур. Опять же, нам нужно проявлять творческий подход и творчески использовать существующие инструменты.

Один из способов нарисовать полосу подписи — добавить текст на темном фоне. Ниже мы добавляем текст ©DATAQUEST Source: P. Cortez et al.с цветом фона #4d4d4d(оттенком серого). Сам текст имеет цвет #f0f0f0, который совпадает с цветом фона графика.

```python
# Assume the rest of the code is written
ax.text(-0.7, -2.9,
        '©Моя ПРОФЕССИЯ  ',
        color = '#f0f0f0',
        backgroundcolor = '#4d4d4d',
        size=12)
```

```python
style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(9, 5))
ax.barh(white_wine.corr()['quality'].index, white_wine.corr()['quality'], left=2, height=0.5)
ax.barh(red_wine.corr()['quality'].index, red_wine.corr()['quality'], height=0.5, left=-0.1)

ax.grid(False)

ax.set_yticklabels([])
ax.set_xticklabels([])

x_coords = {'Алкоголь': 0.82, 'Сульфаты': 0.77, 'pH': 0.91,
            'Плотность': 0.78, 'Общий диоксид серы': 0.59,
            'Свободный диоксид серы': 0.5, 'Хлориды': 0.77,
            'Сахар': 0.81, 'Лимонная кислота': 0.61,
            'Летучая кислотность': 0.5, 'Фиксированная кислотность': 0.4}
y_coord = 9.8
for y_label, x_coord in x_coords.items():
    ax.text(x_coord, y_coord, y_label)
    y_coord -= 1

ax.axvline(0.5, c='grey', alpha=0.1, linewidth=1,
           ymin=0.1, ymax=0.9)
ax.axvline(1.45, c='grey', alpha=0.1, linewidth=1,
           ymin=0.1, ymax=0.9)

ax.axhline(-1, color='grey', linewidth=1, alpha=0.5,
          xmin=0.01, xmax=0.32)
ax.text(-0.7, -1.7, '-0.5'+ ' '*31 + '+0.5',
        color='grey', alpha=0.5)
ax.axhline(-1, color='grey', linewidth=1, alpha=0.5,
           xmin=0.67, xmax=0.98)
ax.text(1.43, -1.7, '-0.5'+ ' '*31 + '+0.5',
        color='grey', alpha=0.5)

ax.axhline(11, color='grey', linewidth=1, alpha=0.5,
           xmin=0.01, xmax=0.32)
ax.text(-0.33, 11.2, 'КРАСНОЕ ВИНО', weight='bold')
ax.axhline(11, color='grey', linewidth=1, alpha=0.5,
           xmin=0.67, xmax=0.98)
ax.text(1.75, 11.2, 'БЕЛОЕ ВИНО', weight='bold')

ax.text(-0.7, -2.9,
        '©Моя ПРОФЕССИЯ' + ' '*94 + 'Аналитик.            ',
        color = '#f0f0f0', backgroundcolor = '#4d4d4d',
        size=12)
```

Последнее изменение, которое мы собираемся внести в наш график, — это цвета полос:

- Положительные значения корреляции примут цвет #33A1C9(оттенок синего).

Чтобы внести это изменение, мы будем использовать colorпараметр Axes.barh(). До сих пор мы использовали только строки с этим параметром: 'blue', 'red', '#af0b1e'и т. д.
