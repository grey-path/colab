## Step 4 Perform the regression

```python
# Add a new column (equal in lenght to x1) consisting only of ones
x = sm.add_constant(x1)

x.head()
```

```python
# Fit the model, according to the OLS (ordinary least squares) method
# where y is a dependent variable and x is an idependent variable
results = sm.OLS(y,x).fit()
```

```python
results.summary()
```

Модель: OLS (Обычный метод наименьших квадратов). OLS — это распространенный метод оценки линейной регрессии. Этот метод находит линию, которая минимизирует сумму квадратов ошибок (более низкая ошибка = лучшая объяснительная способность)

В этом уроке я использую OLS, потому что он довольно простой и достаточно мощный.

R-квадрат измеряет, насколько сильна регрессия

R-квадрат = изменчивость, объясненная регрессией / общая изменчивость набора данных
R-квадрат с использованием значений в диапазоне от 0 до 1, если R-квадрат = 0 означает, что ваша регрессия не объясняет НИ ОДНОЙ изменчивости, если 1 означает, что регрессия объясняет ВСЮ изменчивость

Пример: наш R-квадрат имеет 0,406, другими словами, результаты SAT объясняют 41% изменчивости оценок в колледже, но, поскольку это далеко от 90%, мы можем сделать вывод, что нам не хватает некоторой важной информации. Необходимо учитывать и другие детерминанты, такие как пол, доход или, возможно, семейное положение.

Вывод: R-квадрат измеряет качество подгонки, чем больше факторов вы включаете в регрессию, тем выше R-квадрат.

F-статистика: важна для регрессии, поскольку она дает нам некоторые важные сведения (чем выше значение, тем лучше), чем ниже F-статистика, тем ближе к незначимой модели. Вы можете использовать этот инструмент для сравнения с другими моделями.

```python
plt.scatter(x1,y)
yhat=0.2750+0.0017*x1 #yHat=0.275+0.0017x1 Regression Line
fig=plt.plot(x1,yhat, lw=4,c='orange',label='Regression Line')
plt.xlabel('SAT',fontsize='20')
plt.ylabel('GPA',fontsize='20')
plt.show()
```
