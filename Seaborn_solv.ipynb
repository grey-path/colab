```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')
sns.histplot(tips['total_bill'], kde=True, color='blue')
plt.title('Distribution of Total Bill')
plt.show()
```

2.

```python
sns.scatterplot(x='total_bill', y='tip', hue='sex', data=tips, palette='coolwarm')
sns.regplot(x='total_bill', y='tip', data=tips, scatter=False, color='green')
plt.title('Total Bill vs Tip by Gender')
plt.show()
```

3.

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='Set3')
plt.title('Boxplot of Total Bill by Day of the Week')
plt.show()
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species', palette='muted')
plt.suptitle('Pairplot of Iris Dataset', y=1.02)
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
corr = titanic.corr(numeric_only=True)

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Titanic Dataset')
plt.show()
```

```python
sns.countplot(x='class', hue='sex', data=titanic, palette='Set2')
plt.title('Count of Passengers by Class and Gender')
plt.show()
```

```python
sns.kdeplot(data=titanic, x='fare', hue='class', fill=True, common_norm=False, palette='Set1')
plt.title('Fare Distribution by Class')
plt.show()
```

```python
sns.lmplot(x='age', y='fare', hue='sex', col='class', data=titanic, palette='coolwarm', aspect=0.7)
plt.suptitle('Age vs Fare with Regression Lines by Sex and Class', y=1.02)
plt.show()
```

```python
g = sns.FacetGrid(tips, col='time', row='sex', margin_titles=True, height=4)
g.map(sns.regplot, 'total_bill', 'tip')
g.fig.suptitle('Total Bill vs Tip by Time of Day and Gender', y=1.03)
plt.show()
```
