## PYTHON | PANDAS.MELT ()

Чтобы упростить анализ данных в таблице, мы можем преобразовать данные в более удобную форму, используя Pandas в Python. <br>__Pandas.melt ()__ — одна из функций, которая делает это ..<br>
__Pandas.melt ()__ разворачивает DataFrame из широкого формата в длинный.<br>
Функция __melt ()__ полезна для преобразования DataFrame в формат, в котором один или несколько столбцов являются переменными идентификатора, тогда как все остальные столбцы, считающиеся измеренными переменными, не поворачиваются к оси строк, оставляя только два столбца без идентификатора, переменную и значение.<br>
Синтаксис:

pandas.melt(frame, id_vars=None, value_vars=None,
 var_name=None, value_name='value', col_level=None)

```python
df = pd.DataFrame({'Name': {0: 'John', 1: 'Bob', 2: 'Shiela'},

                   'Course': {0: 'Masters', 1: 'Graduate', 2: 'Graduate'},

                   'Age': {0: 27, 1: 23, 2: 21}})

df

```

```python
# Имя - id_vars, а курс - value_vars.

pd.melt(df, id_vars =['Name'], value_vars =['Course'])
```

```python
# несколько столбцов отмены

pd.melt(df, id_vars =['Name'], value_vars =['Course', 'Age'])
```

```python
# Имена столбцов переменных и значений могут быть изменены

pd.melt(df, id_vars =['Name'], value_vars =['Course'],

              var_name ='ChangedVarname', value_name ='ChangedValname')
```

__Задача 1:__ Написать функцию, которая принимает значение rate и анализирует population. Обратите внимание,что  population находится на правой стороне /.

```python
tbl3.dtypes
```

поскольку скорость является «object», это строка string

```python
def extract_population(rate):
    population = rate.split('/')[1]
    return population
```

```python
#быстрый способ убедиться, что функция не выйдет из строя

assert extract_population("123/456") == "456"
```

```python
assert extract_population("123/456") == "123"
```

```python
# применим функцию к столбцу

tbl3["rate"].apply(extract_population)
```

__Задача 2:__ Установить population в новый столбец.

```python
pop_column = tbl3["rate"].apply(extract_population)
```

```python
tbl3["population"] = pop_column
```
